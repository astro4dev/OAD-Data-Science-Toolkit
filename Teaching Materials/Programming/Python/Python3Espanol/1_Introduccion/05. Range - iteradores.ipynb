```python
# esta celda funciona en binder
list(range(10))
```

```python
import turtle   
ventana = turtle.Screen()     
german = turtle.Turtle()
```

```python
for i in range(10):
    german.forward(15)
    german.right(90)
    german.forward(15)
    german.left(90)
```

```python
repeticiones=10
dist=15
angulo=90
for i in range(repeticiones):
    german.forward(dist)
    german.right(angulo)
    german.forward(dist)
    german.left(angulo)
```

```python
# esta celda funciona en binder
ar=range(0, 19, 2)
br=range(0, 20, 2)
cr=range(10, 0, -1)
```

```python
# esta celda funciona en binder
for a in ar:
    print(a) # inténtalo con los demás rangos, br y cr para ver qué ocurre
```
